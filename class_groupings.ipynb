```python
features = data[['ClassParticipation', 'ParentalInvolvement',
       'CourseCompletion', 'assignemnts score', 'sba score', 'mid-term score']]
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(features)
```

```python
features.head()
```

```python
model.labels_
```

```python
#passing in new data to predict
model.predict([[3,4,2,80,80,80]])
```

```python
model.cluster_centers_
```
